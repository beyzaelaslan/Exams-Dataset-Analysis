#### statistics of the numeric columns visualization with bar graph

```python
visual_summary = df.describe()

for column in visual_summary.columns:
    plt.bar(visual_summary.index, visual_summary[column])
    plt.title(f'{column} Summary')
    plt.ylabel(column)
    plt.show()
```

### mean and median representation usign histogram 

```python

mean_writing_score = df['writing score'].mean()
median_writing_score = df['writing score'].median()

sns.histplot(df['writing score'], color='grey')
#emphasizing a median and mean as a x value
plt.axvline(mean_writing_score, color='blue',  linewidth=2, label=f'Mean: {mean_writing_score}')
plt.axvline(median_writing_score, color='orange' ,linewidth=2, label=f'Median: {median_writing_score}')
plt.title('Histogram of Writing Score')
plt.xlabel('Writing Score')
plt.legend()
plt.show()


```

### mode representation of the genders in pie chart

```python
mode_gender = df['gender'].mode()
#gender columunda cinsiyeti ka kez tekrarlandn belirler
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%.1f%%', startangle=90, colors=['green', 'pink'])
plt.title('Pie chart of the Genders')
plt.show()
```

###  count of parents categorised in the  level of education

```python
parental_education_values = df['parental level of education'].value_counts()
sns.barplot(x=parental_education_values, y=parental_education_values.index)
plt.title('Parent Educational Level')
plt.xlabel('Count')
plt.show()
```

### analysis of  how many students completed the preparation course

```python
completion_counts = df['test preparation course'].value_counts()

sns.barplot(x=completion_counts.index, y=completion_counts.values)
plt.title('Completion Status of the Preparation Course')
plt.xlabel('Completion Status')
plt.ylabel('Number of Students')

# this for loop is use for print actual numbers of the studentr
# +5 is for the location adjustment ( it adds space between bars and the count)
for i, count in enumerate(completion_counts.values):
    plt.text(i, count +5, str(count))

plt.show()
```
